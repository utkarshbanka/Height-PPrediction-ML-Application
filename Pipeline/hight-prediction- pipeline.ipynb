```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('dataclean.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['height']) 
y = df['height']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train = x_train.values
x_train
```

```python
x_test = x_test.values
x_test
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
lasso = Lasso(alpha=6.5)  
lasso.fit(x_train, y_train)
```

```python
lasso_test_pred = lasso.predict(x_test)
lasso_train_pred = lasso.predict(x_train)
```

```python
train_r2 = r2_score(y_train, lasso_train_pred)
train_mae = mean_absolute_error(y_train, lasso_train_pred)
train_mse = mean_squared_error(y_train, lasso_train_pred)

# Test metrics
test_r2 = r2_score(y_test, lasso_test_pred)
test_mae = mean_absolute_error(y_test, lasso_test_pred)
test_mse = mean_squared_error(y_test, lasso_test_pred)

print("Train R2:", train_r2)
print("Train MAE:", train_mae)
print("Train MSE:", train_mse)

print("Test R2:", test_r2)
print("Test MAE:", test_mae)
print("Test MSE:", test_mse)
```

```python
import pickle

LS = lasso
with open("model.pkl", "wb") as f:
    pickle.dump(LS, f)
```
